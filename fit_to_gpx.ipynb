```python
activites_data_fit = "strava_archive/activities/*.json"
```

```python
columns_dict = {'latitude':'lat','longitude':'lon', 'altitude':'ele', 'timestamp':'time'}
columns = list(columns_dict.values())+['filename']
```

```python

glob.glob(activities_data_fit
```

```python
full_df = pd.DataFrame()

for fit_file in tqdm(glob.glob(activities_data_fit)):
    _, tmp_df = conv.fit_to_dataframes(fit_file)
    if tmp_df.shape[0] == 0:
        continue
    tmp_df['filename'] = fit_file
    tmp_df = tmp_df.rename(columns = columns_dict)[columns].fillna(0)
    full_df = pd.concat([full_df, tmp_df], axis=0)
```

```python
import pickle
```

```python
pickle.dump(full_df, open('fit2gpx_df.pkl','wb'))
```
